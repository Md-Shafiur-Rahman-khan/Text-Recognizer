```python
from collections import Counter

# Function to check if word can be spelled with given characters
def checkword(word, chars):
    char_counter = Counter(chars)  # Count occurrences of each character in chars
    for c in word:
        if char_counter[c] > 0:  # If character exists and has available count
            char_counter[c] -= 1  # Decrease the count for the character
        else:
            return False  # Return False if character is not available or exhausted
    return True  # All characters in word were found in chars
```

```python
import pandas as pd
from collections import Counter
import random  # To shuffle the results

# Step 1: Define the checkword function to check if a word can be spelled with the given characters
def checkword(word, chars):
    """
    Checks if the word can be constructed using the characters in 'chars'.
    Returns True if the word is valid, False otherwise.
    """
    # Count occurrences of each character in 'chars'
    char_counter = Counter(chars)
    
    # Check if each character in the word can be matched in 'chars'
    for c in word:
        if char_counter[c] > 0:  # If character exists and has available count
            char_counter[c] -= 1  # Decrease the count for that character
        else:
            return False  # Return False if character is not available or exhausted
    
    return True  # Return True if the word is valid

# Step 2: Create a sample DataFrame of words
df_words = pd.DataFrame({
    'word': ['hello', 'world', 'python', 'pi', 'mirror', 'rope', 'prime', 'iron']
})

# Step 3: Define available characters (in lowercase)
chars = "rorpempii"

# Step 4: Apply the checkword function to each word in the DataFrame
df_words['valid'] = df_words['word'].apply(lambda word: checkword(word, chars.lower()))

# Step 5: Filter the DataFrame to get only valid words
df_sol = df_words[df_words['valid']]

# Step 6: Shuffle the results to randomize the order of valid words
df_sol = df_sol.sample(frac=1).reset_index(drop=True)  # Randomize and reset index

# Step 7: Print the resulting DataFrame with valid, shuffled words
print(df_sol)
```

```python
import pandas as pd
import random
import ipywidgets as widgets
from IPython.display import display

# Step 1: Define the checkword function to check if a word can be spelled with the given characters
def checkword(word, chars):
    """
    Checks if the word can be constructed using the characters in 'chars'.
    Returns True if the word is valid, False otherwise.
    """
    char_counter = Counter(chars)
    for c in word:
        if char_counter[c] > 0:
            char_counter[c] -= 1
        else:
            return False
    return True

# Step 2: Sample DataFrame of words (for demonstration purposes)
df_words = pd.DataFrame({
    'word': ['hello', 'world', 'python', 'pi', 'mirror', 'rope', 'prime', 'iron'],
    'definition': ['A greeting', 'The earth', 'A programming language', 'Mathematical constant', 'A reflective surface', 'A type of string', 'A prime number', 'Metal'],
    'frequency': [3, 1, 2, 1, 2, 1, 1, 2]  # Example frequency of appearance
})

# Step 3: Define available characters
chars = "rorpempii"

# Step 4: Apply the checkword function to each word in the DataFrame to determine if it is valid
df_words['valid'] = df_words['word'].apply(lambda word: checkword(word, chars.lower()))

# Step 5: Add additional columns: 'length' and 'frequency' (already included, just for the purpose of this example)
df_words['length'] = df_words['word'].apply(len)

# Step 6: Filter the DataFrame to get only valid words
df_sol = df_words[df_words['valid']]

# Step 7: Shuffle the results to randomize the order of valid words
df_sol = df_sol.sample(frac=1).reset_index(drop=True)

# Step 8: Create an interactive table using widgets (for demonstration)
def show_interactive_table(df):
    """
    Function to display the DataFrame in an interactive table format.
    """
    # Create an interactive widget for the DataFrame display
    table_widget = widgets.Output()

    with table_widget:
        # Display the DataFrame
        display(df)

    return table_widget

# Step 9: Display the interactive table
interactive_table = show_interactive_table(df_sol.loc[:, ['word', 'length', 'frequency', 'definition']])
display(interactive_table)
```
